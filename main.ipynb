```python
## Dictionaries and Lists ##

cleanTargetDict = {
                    "of something that is not listed here": "Other",
                    "it's too far / too many lanes to cross": "people have to cross too many lanes / too far",
                    "the roadway surface needs maintenance": "the roadway surface needs improvement",
                    'the wait for the "Walk" signal is too long' : 'people have to wait too long for the "Walk" signal',
                    "there's not enough time to cross the street" : "people are not given enough time to cross the street",
                    "there are no sidewalks or they need maintenance" : "sidewalks/ramps don't exist or need improvement",
                    "bike facilities don't exist or need improvement" : "there are no bike facilities or they need maintenance",
                    "it’s hard to see / low visibility" : "it’s hard for people to see each other"
}

```

```python
def loadData(clean=True):
    # Data Import and Load
    url        = 'http://bostonopendata-boston.opendata.arcgis.com/datasets/5bed19f1f9cb41329adbafbd8ad260e5_0.csv'
    df = pd.read_csv(url)
    dropColumns = [col for col in df.columns if col not in ['REQUESTTYPE','COMMENTS']]
    df = df.drop(dropColumns,axis=1)
    if clean == True:
        df = cleanTarget(df) 
        df = cleanTarget2(df)
        df = df[pd.notnull(df.COMMENTS)]  ## Dropping Rows with No Comment
    return df

def cleanTarget(df):
    clean_requests = []
    for request in df.REQUESTTYPE:
        try:
            garbage,real_request=request.split(';')
            clean_requests.append(real_request)
        except:
            clean_requests.append(request)
    df['REQUESTTYPE']=clean_requests
    return df

def cleanTarget2(df):
   
    for key,value in cleanTargetDict.items():
        df.REQUESTTYPE = df.REQUESTTYPE.replace(key,value,regex=True)
    return df

def cleanTarget3(df):
    ## Replace REQUESTTYPES with numbers
    df2 = df.copy()
    i = 1
    trackList={}
    for item in df2.REQUESTTYPE.unique():
        trackList[i] =item
        df2.REQUESTTYPE = df2.REQUESTTYPE.replace(item,i,regex=True)
        i+=1
    return df2,trackList

#Feature Engineering. Basically creating 4 different features based on Comments
def cleanComment(df):
    df3=df.copy().drop_duplicates('COMMENTS')
    df3["NWords"] = df3["COMMENTS"].apply(lambda x: len(x.split(" ")))
    df3["NChars"] = df3["COMMENTS"].apply(len)
    df3['NExclamations'] = df3['COMMENTS'].apply(lambda x: len(x.split('!'))-1)
    df3['NQMarks'] = df3['COMMENTS'].apply(lambda x: len(x.split('?'))-1)
    df3 = df3[df3.NWords> 1]
    df3 = df3[df3.COMMENTS !='TEST TEST TEST']
    df3 = df3[df3.COMMENTS !='Test. Doit GIS. Test on mobile']
    return df3
    
def getNaN(df):
    print(df.isnull().sum())
    return


def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = np.round(cm.astype('float') / cm.sum(axis=1)[:, np.newaxis],decimals=1)
#         print("Normalized confusion matrix\n")
#     else:
#         print('Confusion matrix, without normalization\n')

#     print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    return

def plot_important_features(coef, feature_names, top_n=20, ax=None):
    
    if ax is None:
        ax = plt.gca()
    inds = np.argsort(np.abs(coef))[-top_n:]
    important_coefs = coef[inds]
    sort_sign = np.argsort(important_coefs)
    myrange = range(len(inds))
    combined_inds = inds[sort_sign]

    ax.bar(myrange, coef[combined_inds])
    ax.set_xticks(myrange)
    ax.set_xticklabels(feature_names[combined_inds], rotation=60, ha="right")
    return feature_names[combined_inds]
```

## Task2 - Model 1


Run a baseline multi-class classification model using a bag-of-word approach, report macro f1-score (should be above .5) and visualize the confusion matrix. Can you interpret the mistakes made by the model? 

```python
#creating training and test datasets
X = boston_fewfeatures2.copy() 
Y = pd.DataFrame(X['REQUESTTYPE'].copy())
del X['REQUESTTYPE']

#getting train and test datasets
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,stratify=Y,random_state=0)
print(X_train.shape)
print(Y_train.shape)
```

```python
#Scaling the new features using MaxAbsScaler
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
for f in X_train.columns:
        if X_train[f].dtype!='object':
            scale = MaxAbsScaler()
            X_train_scaled[f] =  scale.fit_transform(X_train[f])
            X_test_scaled[f] = scale.transform(X_test[f])
```

```python
'''Baseline multi class classification model using Bag of Word approach and Logistic Regression'''
np.random.seed(0)

vect=CountVectorizer()
sparse_M_train=vect.fit_transform(X_train_scaled['COMMENTS'].copy())
sparse_M_test=vect.transform(X_test_scaled['COMMENTS'].copy())
feature_names=vect.get_feature_names()


NumCols = ['NWords','NChars']

X_train_vect = sparse.hstack([X_train_scaled[NumCols], sparse_M_train]).tocsr()
X_test_vect = sparse.hstack([X_test_scaled[NumCols], sparse_M_test]).tocsr()


lr=LogisticRegression(C=0.2).fit(X_train_vect,Y_train)
predict=lr.predict(X_test_vect)
f1_score(Y_test,predict,average='macro')
```

```python
print(classification_report(Y_test,predict))
```

```python
##### Exploratory Grid Search for C of Logistic Regression  ####

np.random.seed(0)
alpha  = np.arange(0.1,0.25,0.01)
accuracy, fscore, precision, recall = [],[],[],[]

for est in alpha:
    clf = LogisticRegression(C =est).fit(X_train_vect, Y_train)
    y_pred = clf.predict(X_test_vect)
    accuracy.append(accuracy_score(Y_test, y_pred))
    fscore.append(f1_score(Y_test, y_pred, average='macro')) 

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 18
fig_size[1] = 12

plt.plot(alpha, accuracy, label="accuracy scores",color='b')
plt.plot(alpha, fscore, label="fscore_binary scores",color='r') 
plt.xlabel("estimators")
plt.ylabel("score")
plt.legend()
plt.show()

```

```python
'''CONFUSION MATRIX'''
class_names = Y['REQUESTTYPE'].unique()
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 16
plt.rcParams.update({'font.size':12})

# Compute confusion matrix
cnf_matrix = confusion_matrix(Y_test,predict)

print("\t\t\t\t Class Dictionary Pairs (REQUESTTYPE) Keys with Values\n\t\t\t\t\tFor Reference to Confusion Matrix\n")
for key, value in trackList.items():
    print("\t\t\t\t",key,": ", value)

# # Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix, without normalization',cmap=plt.cm.Blues)
plt.show()

print("\t\t\t\t\t\n","*"*100,"\n")

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix', cmap=plt.cm.coolwarm)

plt.show()
```
